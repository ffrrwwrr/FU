```python
# 4. Удалить из набора поля, не несущие информативной нагрузки - содержащие уникальные либо константные значения
data.drop(columns = ['Name'], axis = 1)
```

```python
# 5. Заменить возраст пассажиров на год рождения
df5 = pd.DataFrame(data, columns=['Name', 'PClass', 'Age', 'Sex', 'Survived'])
df5.loc[(df5.Age > 0), 'Age'] = round(1912 - df5["Age"])
df5
```

```python
# 6. Переименовать все названия колонок на русском языке
df6 = pd.DataFrame(data, columns=['Name', 'PClass', 'Age', 'Sex', 'Survived'])
df6.rename(columns = {'Name' : 'Имя', 'PClass' : 'Класс', 'Age' : 'Возраст', 'Sex' : 'Пол', 'Survived' : 'Выжил'}, inplace = True)
df6
```

```python
# 7. Заменить все отсутствующие данные средним значением (в случае числового поля) 
#    либо самым распространенным в категориальных полях.
print (f'''Кол-во отсутсвующих данных в каждом поле:\n{data.isnull().sum()}\n''')

df7 = pd.DataFrame(data, columns=['Name', 'PClass', 'Age', 'Sex', 'Survived'])
df7["Age"].fillna(df7["Age"].mean(), inplace = True) # заполнили пустые клетки в столбце 'Age' средним по столбцу через .mean()
df7["PClass"].fillna(df7["PClass"].mode()[0], inplace = True) # заполнить пустые клетки в столбце 'PClass' через Моду .mode()
df7.rename(columns = {'Name' : 'Имя', 'PClass' : 'Класс', 'Age' : 'Возраст', 'Sex' : 'Пол', 'Survived' : 'Выжил'}, inplace = True)
df7.drop(columns = ['Имя'], axis = 1, inplace = True)
df7
```

```python
# 8. Сохранить получившийся набор данных в формате CSV
df7.to_csv('task08.csv', encoding='WINDOWS-1251', index = False)
```

```python
# 9. Открыть данный файл в программе Microsoft Excel и сохранить в формате .xlsx
# 10. Считать файл .xlsx в новый DataFrame
pd.read_excel('task08.xlsx') 
```

```python
# 6. Напишите участок кода, демонстрирующий возможность добавления новой строки к набору данных.
import pandas as pd
data6 = {
    'name': ['Katy', 'Max', 'Artem', 'Lisa'], 
    'age': [15, 25, 23, 45], 
    'sex': ["W", "M", "G", "W"]
        }


df6 = pd.DataFrame(data6)
print(f'До добавления новой строки:\n\n{df6}\n\nПосле добавления новой строки:')
df6.append({'name' : 'Farkhod', 'age' : 20, 'sex' : "M"}, ignore_index=True)
```

```python
# 7. Напишите участок кода, демонстрирующий возможность добавления новой колонки (поля) к набору данных.
data7 = {
    'name': ['Katy', 'Max', 'Artem', 'Lisa'], 
    'age': [15, 25, 23, 45], 
    'sex': ["W", "M", "G", "W"]
        }

df7 = pd.DataFrame(data7)
print(f'До добваления нового столбца:\n\n{df7}\n\nПосле добваления нового столбца:')
size = ['M', 'S', 'L', 'XL']
df7["size"] = size
df7
```
