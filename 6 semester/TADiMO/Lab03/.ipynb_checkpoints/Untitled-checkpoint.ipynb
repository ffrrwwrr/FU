```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1
x = np.linspace(-4, 4, 100)
y = x**2 + x + 1
plt.title('График')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(linestyle = '--')

plt.annotate('Точка максимума', xy=(4,21),  xycoords='data',
            xytext=(0, 20), textcoords='data',
            arrowprops=dict(arrowstyle="->", facecolor='black'))

plt.scatter(4,21, color='black', s=50)
plt.plot(x, y)
plt.show()
```

```python
#2
x = np.linspace(-4,4, 100)
y = (x/np.log(x))
plt.title('График')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(linestyle = '--')
plt.plot(x, y)
plt.show()
```
