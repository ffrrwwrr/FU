### Объединение таблиц `pd.DataFrame`

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты без отзывов должны отсутствовать в данной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и выведя на экран строку из полученного `DataFrame`, содержащую информацию об этом отзыве.

```python
t16 = pd.read_csv("data/recipes_sample.csv")
t16_1 = pd.read_csv("data/reviews_sample.csv")

# Таблица рецептов с отзывами
t16_2 = t16.merge(t16_1, left_on = 'id', right_on = 'recipe_id')[["id", "name", "user_id", "rating"]]
print(f"DataFrame 1:\n{t16_2}")

# Таблица рецептов без отзывов
t16_4 = t16.merge(t16_1, left_on = 'id', right_on = 'recipe_id', how = 'left')
t16_5 = t16_4[pd.isnull(t16_4['user_id'])][["id", "name", "user_id", "rating"]]
print(f"\n------------------------------------------------------------------------\n\nDataFrame 2:\n{t16_5}")

# Проверка таблицы 1
print(f"\n------------------------------------------------------------------------\n\nCheck:\n{t16_2[t16_2['id'] == 223349]}")
```

5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`. У рецептов, для которых отсутствуют отзывы, в соответствущем столбце должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и выведя на экран строку из полученного `DataFrame`, содержащую информацию об этом отзыве.


```python
t17 = t16.merge(t16_1, left_on = 'id', right_on = 'recipe_id', how = 'left').groupby('id').count()['review'].reset_index()
t17_1 = t16[['id', 'name']].merge(t17, on = 'id').rename(columns={'id': 'recipe_id', 'review': 'review_count'})
print(f"{t17_1}\n\n------------------------------------------------------------------------\n\nCheck:\n{t17_1[t17_1['review_count'] == 0]}")
```

5.3. Выясните, отзывы, добавленные в каком году, имеют наименьший средний рейтинг?

```python
t18 = t16_1['rating'].groupby(pd.to_datetime(t16_1['date']).dt.year).mean().reset_index()
print(f"DataFrame:\n{t18}\n\n------------------\n\nAnswer:\n{t18[t18['rating'] == t18['rating'].min()]}")
```
